# Phone Usage In India

# The dataset represents simulated phone usage data for Indian users. It contains 17,686 rows and 16 columns, each row representing an individual's phone usage details. The data reflects various aspects of mobile phone behavior, including demographics, phone brand preferences, daily screen time, and app usage patterns.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('phone_usage_india.csv')
```

# Location that has more phone users

```python
plt.figure(figsize=(5,5))

sns.countplot(data=df,x='Location',color='green')
plt.xticks(rotation=90)
plt.show()
```

```python
df.Location.value_counts()
```

# Genderwise has more users

```python
ax=sns.countplot(data=df,x='Gender')
ax.bar_label(ax.containers[0])
```

# Phone brands prefer By users

```python
plt.figure(figsize=(10, 5))
ax=sns.countplot(data=df, x='Phone Brand', color='red')
ax.bar_label(ax.containers[0])
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.title('Phone Brand Count')
plt.show()
```

# Locaton wise more phone users

```python
plt.figure(figsize=(14, 6))
sns.countplot(y=df['Location'], order=df['Location'].value_counts().index, palette='viridis')
plt.title('Countplot of Location', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Location', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

# Gender Vs Call duration (mins/day)

```python
plt.figure(figsize=(5, 5))

ax=sns.barplot(x='Gender', y='Calls Duration (mins/day)', data=df, estimator='mean', palette='muted', ci=None)
ax.bar_label(ax.containers[0])
plt.title('Gender Vs Call durartion')
```

```python
plt.figure(figsize=(5, 5))
df.groupby('Gender')['Age'].mean().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('CustomerOccupation Distribution by AverCustomerAge CustomerAge', fontsize=16)
plt.ylabel('')  # Remove default y-axis label
plt.show()
```

# Gender Vs Screen Time hrs/day

```python
screen_time_by_gender = df.groupby('Gender')['Screen Time (hrs/day)'].sum().reset_index()
plt.figure(figsize=(5,5))
ax = sns.barplot(data=screen_time_by_gender, x='Gender', y='Screen Time (hrs/day)', palette='muted')
ax.bar_label(ax.containers[0])
plt.title('Total Screen Time by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Total Screen Time (hrs/day)', fontsize=12)
plt.show()
```

# Gender Vs Calls Duration (mins/day)

```python
CallDuration_time_by_gender = df.groupby('Gender')['Calls Duration (mins/day)'].sum().reset_index()
plt.figure(figsize=(5,5))
ax = sns.barplot(data=CallDuration_time_by_gender, x='Gender', y='Calls Duration (mins/day)', palette='muted')
ax.bar_label(ax.containers[0])
plt.title('Total call duration by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Calls Duration (mins/day', fontsize=12)
plt.show()
```

# Primary Use Vs Monthly Recharge Cost (INR)

```python
Monthly_recharge_on_usage = df.groupby('Primary Use')['Monthly Recharge Cost (INR)'].sum().reset_index()

plt.figure(figsize=(10,5))
ax = sns.barplot(data=Monthly_recharge_on_usage, x='Primary Use', y='Monthly Recharge Cost (INR)', palette='muted')

ax.bar_label(ax.containers[0])

plt.title('Monthly Recharge Cost by Primary Use', fontsize=14)
plt.xlabel('Primary Use', fontsize=12)
plt.ylabel('Monthly Recharge Cost (INR)', fontsize=12)

plt.xticks(rotation=45)

plt.show()
```
